```python
#import the neccesary libraries.
import pandas as pd
```

```python
#Creating three data frames based on the following data.
raw_data_1 = {
    'subject_id': ['1', '2', '3', '4', '5'],
    'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
    'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
raw_data_2 = {
    'subject_id': ['4', '5', '6', '7', '8'],
    'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
    'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
raw_data_3 = {
    'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
    'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

```python
#assign each to a variable called data1, data2, data3.
data1 = pd.DataFrame(raw_data_1)
data2 = pd.DataFrame(raw_data_2)
data3 = pd.DataFrame(raw_data_3)

print(data1)
```

```python
#Join the dataframes across rows and assign all data.
pd.concat([data1,data2,data3], ignore_index = True, axis = 1)
```

```python
#Join the two data frames alng columns and assing to all_data_col
pd.concat([data1,data2,data3], ignore_index = True, axis = 0)
```

```python
#print data3
print(data3)
```

```python
#merge all data and data3 along the subject_id value.
all_data = data1.merge(data2, on = 'subject_id', how = 'outer')
all_data
```

```python
all_data.merge(data3, on = "subject_id", how = "outer")
```

```python
#merge only the data that has the same subject_id of both data1 and data2
data1.merge(data2, on = "subject_id", how = "inner")
```

```python
#merge all the values in data1 and data2 with matching records from both sides where available.
data1.merge(data2, on = "subject_id", how = "inner")
```
