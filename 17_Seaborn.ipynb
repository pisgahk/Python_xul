```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
#Load an example dataset
data = sns.load_dataset('fmri')

#creating a line plot.
sns.lineplot(x='timepoint',y='signal', hue='region', style='event', data=data)

#show the plot
plt.title('line plot example')
plt.show()
```

```python
import warnings

#suppress the specific Future warning from seaborn
warnings.filterwarnings("ignore", category=FutureWarning, module='seaborn')

#load an example dataset.
data = sns.load_dataset('fmri')

#create a line plot
sns.lineplot(x='timepoint', y='signal', hue='region', style='event', data=data)

#show the plot
plt.title('Line Plot Example')
plt.show()
```

```python
titanic = sns.load_dataset('titanic') #Load an example dataset.

sns.barplot(x='class', y='survived', hue='sex', data=titanic)  #create a bar plot.

plt.title('Bar plot example')
plt.show()
```

```python
iris = sns.load_dataset('iris')

sns.histplot(iris['sepal_length'], bins=10, kde=True)

plt.title('Histogram Example.')
plt.show()
```

```python
# Density Plots
iris = sns.load_dataset('iris')

sns.kdeplot(iris['sepal_length'], fill=True)

plt.title('Density Plot Example.')
plt.show()
```

```python
#Scatter plots are used to show the relationship between two variables.
iris = sns.load_dataset('iris')

sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)

plt.title('Scatter Plot Example.')
plt.show()
```

```python
#point plots show point estimates and confidence intervals using scatter points.
titanic = sns.load_dataset('titanic')

sns.pointplot(x='class', y='survived', hue='sex', data=titanic)

plt.title('Point Plot Example.')
plt.show()
```

```python
#facet grids
tips = sns.load_dataset('tips')

g = sns.FacetGrid(tips, col='time', row='sex')
g.map(sns.scatterplot, 'total_bill', 'tip')

plt.title("Facet Grids")
plt.show()
```

```python
#categorial data
titanic = sns.load_dataset('titanic')

sns.boxplot(x='class', y='age', hue='sex', data=titanic) #create a box plot for categorial data

plt.title('categorial data example')
plt.show()
```
