# `loc` and `iloc`

```python
# 1. loc
# allows indexing and slicing that always references the explicit index.
df.loc[1]
```

```python
# 2. iloc
# allows indexing and slicing that always references the impplicit Python style index.
df.iloc[1]
```

```python
df.iloc[1:4] #iloc is exclusive. #Implicit -> Defined by the system
```

```python
df.loc[1:4] #but loc is inclusive. Explicit -> Defined by the User.
```
