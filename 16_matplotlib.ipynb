```python
fig = plt.figure()

# to add sub plots in a 2x2 grid.
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

#this is to add the values to the graph.
ax1.plot([0,1,2], [0,1,4])
ax2.plot([0,1,2], [0,-1,-4])
ax3.plot([0,1,2], [0,2,1])
ax4.plot([0,1,2], [0,-2,1])

plt.show()
fig.savefig("figure_1.png") #Just saves the image with the specified file format in the cwd.
```

```python
#A line plot is used to represent data points connected by straight line.
#It is commonly used showing trends over a continuous interval.
x = [1,2,3,4,5]
y = [2,4,6,1,3]
plt.plot(x,y)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot")
plt.show()
```

```python
# Bar plot or Bar Chart is used to represent categorial data with 
categories = ['A','B','C','D']
values = [10,18,8,38]
plt.bar(categories, values)

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar plot")
plt.show()
```

```python
cities = ['Ambitka Nagar','Bhopal','Chattisgargh','Delhi']
ppln = [10646,1854,8254,3528]
plt.bar(cities, ppln)

plt.xlabel("Cities")
plt.ylabel("Population")
plt.title("Bar graph showing variation of 'Population' with 'Cities'.")
plt.show()
```

```python
# A scatter plot is used to  visualize the relationship between two variables. 
# Each point on the plot represents a data point with two variables.

x = [10,20,30,40,50]
y = [20,40,20,60,40]
plt.scatter(x,y)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.show()
```

```python
# Pie Chart

labels = ['A','B','C','D']
sizes = [150,300,250,200]

plt.pie(sizes, labels=labels, autopct='%.3f%%')
plt.title("Pie Chart Example")
plt.show()
```

```python
# Setting Colours
plt.plot([1,2,3], [4,5,6], color='blue')
plt.plot([1,2,3], [6,5,4], color='#FF3748') #Hex Colour code.
plt.plot([3,5,3], [8,3,5], color='magenta')
plt.show()
```

```python
#Changing line styles.
plt.plot([1,2,3], [4,5,6], linestyle='-', color='blue')
plt.plot([1,2,3], [6,5,4], linestyle='--', color='#FF3748')
plt.show()
```

```python
#Customising markers.
plt.plot([1,2,3], [4,5,6],marker='D', linestyle='-', color='blue') #'o' for the circle.
plt.plot([1,2,3], [6,5,4],marker='*', linestyle='--', color='#FF3748') #'D' for diamond. 'x' for cross.
plt.show()
```

```python
#customising ticks
plt.plot([1,2,3], [4,5,6])
plt.xticks([1,2,3], ['One', 'Two', 'Three']) #This just labels the respective ordinates.
plt.yticks([4,5,6], ['Four', 'Five', 'Six'])

plt.xlabel("X-Axis label")
plt.ylabel("Y-Axis label")
plt.title("Title of the plot")

plt.show()
```

```python
#Adding Legends
plt.plot([1,2,3], [4,5,6], label='Line A')
plt.plot([1,2,3], [6,5,4], label='Line B')

plt.legend()
plt.show()
```

```python
#Adding annotations to the points on the plot.
plt.plot([1,2,3], [4,5,6])
plt.annotate('Point 1', xy=(1,4), xytext=(1,5), arrowprops=dict(facecolor='red', shrink=0.05))
plt.show()
```

```python
# A figure in matplotlib that can contain multiple subplots 

#create a figure with 2x2 subplots.
fig, axs = plt.subplots(2,2,figsize=(10,10))

#Plotting on each subplot with sifferent colors.
axs[0,0].plot([1,2,3], [1,4,0], 'r') #red plot
axs[0,0].set_title('Subplot 1')

axs[0,1].plot([1,2,3], [1,2,3], 'g') #green plot
axs[0,1].set_title('Subplot 2')

axs[1,0].plot([1,2,3], [9,4,1], 'b') #blue plot
axs[1,0].set_title('Subplot 3')

axs[1,1].plot([1,2,3], [2,2,2], 'k') #black plot
axs[1,1].set_title('Subplot 4')

#red -> 'r'
#blue -> 'b'
#green -> 'g'
#cyan -> 'c'
#magenta -> 'm'
#yellow -> 'y'
#black -> 'k'
#white -> 'w'

plt.tight_layout()
plt.show()
```

```python
#Changing the background color. This is necessary esp when the colour you want to use for graph is white.
fig, ax = plt.subplots()

fig.set_facecolor('red')
ax.set_facecolor('pink')

#some sample graphs.
ax.plot([0,1,2], [0.75, 0.25, 0.75], 'k', label='Black')
ax.plot([0,1,2], [0.1,0.0,0.1], 'w', label='white', linestyle='--') #white line.
plt.show()

#adding a legend with a dark background.
```

```python
#Example with combined concepts

#setup figure and axes.
fig, ax = plt.subplots()

#plot data with different styles.
ax.plot([1,2,3], [4,5,6], color='blue', marker='o', linestyle='-', label='Line 1')
ax.plot([1,2,3], [6,5,4], color='green', marker='x', linestyle='--', label='Line 2')

#customise ticks.
ax.set_xticks([1,2,3])
ax.set_xticklabels(['one','two','three'])
ax.set_yticks([4,5,6])
ax.set_xticklabels(['four','five','six'])

#add labels and title.
ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')
ax.set_title('Combined Example')

#add legend.
ax.legend()

#add annotation.
ax.annotate('Point 1', xy=(1,4), xytext=(1.5,5), arrowprops=dict(facecolor='black', shrink=0.05))

#show plot.
plt.show()
```
